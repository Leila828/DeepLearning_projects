```
from keras.utils import plot_model
```

```
from keras_bert import get_model, compile_model 
```

```
model= get_model(
    token_num=30000,
    head_num=12,
    transformer_num=12,
    embed_dim=768,
    feed_forward_dim=3072,
    seq_len=512,
    pos_num=512,
    dropout_rate=0.05,
)
```

```
compile_model(model)
```

```
model.summary()
```

```
plot_model(model,to_file='bert.png')
```
